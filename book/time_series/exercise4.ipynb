### Autocovariance function (ACF) and PSD 

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from scipy import signal

%matplotlib inline
```

**Introduction:** 

In this exercise, you will focus on normalized auto-covariance function (ACF) and the power spectral density (PSD), and the auto-regressive moving average (ARMA).

**Background knowledge:** 

The package <code>statsmodels.graphics.tsaplots</code> contains functions for computing ACF and PSD. <code>plot_acf</code> is one of these functions, which create automatically a plot. Regarding the PSD, there is a function from the package of <code>scipy.signal</code> and you need to use the <code>signal.periodogram</code> to calculate the PSD. An alternative way to compute the PSD is based on the least-squares harmonic estimation (LS-HE), which is based on hypothesis testing.

**Exercise:** 

We use the above functions to plot the ACF and PSD of white noise time series. Later we also compute them for the ARMA(p,q) process. We generate a white noise process, similar to that created in [Time series components exercise](https://mude.citg.tudelft.nl/book/time_series/exercise1.html#) ($m=501$). We will see that white noise does not show any temporal correlation. 

We start defining the parameters of the white noise $\epsilon \sim \textbf{N} (\mu=0, \sigma_{\epsilon}^2=1)$. As previously done in the [Time series components exercise](https://mude.citg.tudelft.nl/book/time_series/exercise1.html#), the number of observation is $m=501$ and the time interval is $\Delta t = 1$ s. The sampling rate is chosen equal to $f_s=1$ Hz.

```python
mean1 = 0 
sigma1 = 1
m = 501
time = np.arange(m) 
Fs = 1 

e = np.random.normal(loc = mean1, scale = sigma1, size = m) 
yt = e

# plot the time series
plt.figure()
plt.grid()
plt.plot(time, yt, color='pink')
plt.title(r'White noise $\epsilon$ time series')
plt.ylabel(r'$Y$(t)')
plt.xlabel('Time (s)')
```

We can now plot the normalized auto-covariance function (ACF) of the generated noise.
Look at the time series. Do you see temporal correlation?

```python
ACF = plot_acf(yt, lags=None, alpha=0.05, title=r'ACF of white noise $\epsilon$', color='pink')
plt.grid()
```

We can also plot the PSD of the white noise. 

```python
# plot the white noise PSD
F, PSD = signal.periodogram(yt, fs=Fs, scaling='density', return_onesided=False)

plt.figure()
plt.grid()
plt.plot(F, PSD, color='pink')
plt.title(r'PSD of white noise $\epsilon$')
plt.ylabel('Power PSD')
plt.xlabel('Frequency (Hz)')
```

The PSD values seem not to be frequency dependent: it looks flat indicating that all frequencies have identical contributions, even though there are some peaks. Think of white light, that has similar characteristics.
